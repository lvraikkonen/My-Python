# Excel Sheet Column Title

26进制，从后往前求出每一位的字符，然后反转。利用‘A’的ASCII码加上余数，求出相应字符

```python
def convert2title(num):
    if num <= 0:
        return ""
    title = ""
    while num > 0:
        num -= 1
        mod = num % 26
        num /= 26
        title += chr(ord('A') + mod)
    # reverse
    return title[::-1]
print convert2title(28)
```

# Min Stack

用常数时间，取出栈中最小值。用最小栈保存最小值信息

```python
class MinStack:
    def __init__(self):
        self.data = []
        self.min_data = []
        
    def push(self, item):
        self.data.append(item)
        if not self.min_data or item <= self.min_data[-1]:
            self.min_data.append(item)
    
    def pop(self):
        if self.data:
            item = self.data.pop()
            if item == self.min_data[-1]:
                self.min_data.pop()
    
    def top(self):
        if self.data:
            return self.data[-1]
    
    def get_min(self):
        if self.min_data:
            return self.min_data[-1]

ms = MinStack()
ms.push(2)
ms.push(0)
ms.push(0)
ms.push(3)
print ms.getMin()
ms.pop()
ms.pop()
print ms.getMin()
print ms.min_data
```

# Valid Palindrome

回文，正着念，反着念都一样，只考虑字母和数字

```python
def valid_palindrome(s):
    l, r = 0, len(s) - 1
    while l < r:
        # only digit and alpha from left
        while l < r and not(s[l].isdigit() or s[l].isalpha()):
            l += 1
        # only digit and alpha from right
        while l < r and not(s[r].isdigit() or s[r].isalpha()):
            r -= 1
        if s[l].lower() != s[r].lower():
            return False
        l, r = l + 1, r - 1
    return True

print valid_palindrome('radar.')
```

# String to Integer

```python
def str2int(s):
    s = s.strip()
    end = ""
    for i, val in enumerate(s):
        if val in ['+', '-'] and i == 0:
            end += val
        elif val.isdigit():
            end += val
        else:
            break
    # only + or -
    if end in ['+', '-']:
        end = ""
    end = int(end) if end else 0
    if end > 2147483647: 
        return 2147483647
    if end < -2147483648:
        return -2147483648
    return end

print str2int('  +-4145r')
```

# Valid Parentheses

```python
def valid_parentheses(s):
    push_set = ['(', '{', '[']
    pop_set = [')', '}', ']']
    pair = dict(zip(push_set, pop_set))
    
    stack = []
    for element in s:
        if element in push_set:
            stack.append(pair[element])
        elif element in pop_set:
            if not stack or element != stack.pop():
                return False
    if stack:
        return False
    else:
        return True

print valid_parentheses('[()]')
```
