## The Node Class

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
    def get_data(self):
        return self.data
    
    def get_next(self):
        return self.next
    
    def set_data(self, newdata):
        self.data = newdata
    
    def set_next(self, newnext):
        self.next = newnext
```

Now, create node class contains data part and pointer part.

## The UnorderList Class

The `unordered list` will be built from a collection of nodes, each linked to the next by explicit references. 
Initially when we construct a list, there are no items. 

```python
class Unorderlist:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head == None

    def add_item(self, item):
        """ add item at head of linked list """
        tmp = Node(item)
        tmp.set_next(self.head)
        self.head = tmp

    def size(self):
        current = self.head
        count = 0
        # traverse the linked list
        while current != None:
            count += 1
            current = current.get_next()
        return count

    def search(self, item):
        current = self.head
        found = False
        while current != None and not found:
            if current.get_data() == item:
                found = True
            else:
                current = current.get_next()
        return found

    def remove(self, item):
        current = self.head
        previous = None
        found = False
        # first search item
        while not found:
            if current.get_data() == item:
                found = True
            else:
                previous = current
                current = current.get_next()
        # not found
        if not found:
            return "The item %d not exist in linked list." % item
        else:
            # found at head
            if previous == None:
                self.head = current.get_next()
            else:
                previous.set_next(current.get_next())

    def append(self, item):
        """ adds a new item to the end of the list
            making it the last item in the collection """
        current = self.head
        previous = None
        while current != None:
            previous = current
            current = current.get_next()
        # end
        tmp = Node(item)
        previous.set_next(tmp)

    def index(self, item):
        """ returns the position of item in the list.
            It needs the item and returns the index.  """
        current = self.head
        index = 0
        found = False
        while not found and current != None:
            if item == current.get_data():
                found = True
            else:
                current = current.get_next()
                index += 1
        if found:
            return (item, index)
        else:
            return "Item %d doesn't exist in linked list." % item

    def insert(self, pos, item):
        tmp = Node(item)
        if pos < 0:
            return "Invalid position!"
        elif pos == 0:
            # insert at head
            tmp.set_next(self.head)
            self.head = tmp
        elif pos > self.size():
            # insert at end
            self.append(item)
        else:
            current = self.head
            previous = None
            index = 0
            while index < pos:
                previous = current
                current = current.get_next()
                index += 1
            # found the position
            tmp.set_next(current)
            previous.set_next(tmp)

    def pop(self, pos):
        """ removes and returns the item at position pos. """
        index = 0
        if pos >= self.size()-1:
            # remove from end
            current = self.head
            previous = None
            while current.get_next():
                previous = current
                current = current.get_next()
            result = current.get_data()
            previous.set_next(None)
            return result
        else:
            current = self.head
            while index < pos:
                current = current.get_next()
                index += 1
            # fake remove current with constant time O(1)
            result = current.get_data()
            current.set_data(current.get_next().get_data())
            tmp = current.get_next()
            current.set_next(tmp.get_next())

            return result

    def show_list(self):
        current = self.head
        result = []
        while current != None:
            result.append(current.get_data())
            current = current.get_next()
        print result

my_list = Unorderlist()
my_list.add_item(7)
my_list.add_item(6)
my_list.add_item(5)
print my_list.size()
my_list.show_list()
my_list.append(8)
my_list.show_list()
print my_list.index(7)
my_list.insert(0, 4)
my_list.show_list()
my_list.insert(2, 'a')
my_list.show_list()
print my_list.pop(2)
my_list.show_list()
print my_list.pop(my_list.size())
my_list.show_list()
```
